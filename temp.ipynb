```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression   #gradient assent
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
d={"temp":[20,22,25,27,30,32,35,37,40],"sales":[200,250,300,330,370,400,460,500,550]}
df=pd.DataFrame(d)
df.head()
```

```python
x=df[['temp']]
y=df[['sales']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
model=LinearRegression()
```

```python
#Train the model

model.fit(x_train,y_train)  #internally it runs y=mx+c and give one equation
```

```python
model.score(x_train,y_train)*100,model.score(x_test,y_test)*100   #for accuracy check
```

```python
model.intercept_
```

```python
model.coef_    #slope
```

```python
ypred=model.predict(x_test)
ypred  #we use y_test because machine already now the train datas
```

```python
y_test-ypred
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x_test,y_test)
plt.scatter(x_test,ypred)
plt.show()
```

```python
r2_score(y_test,ypred)   #performance check
```

```python
model.score(x_test,y_test)         
```

```python
mean_squared_error(y_test,ypred)    #to find average error difference here we do squersfre now so we get more value so at that stuaton we use MAE
```

```python
mean_absolute_error(y_test,ypred)
```
